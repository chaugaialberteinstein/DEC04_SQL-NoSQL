```
import pandas as pd
import matplotlib.pyplot as plt
```

```
df = pd.read_csv('task4.2_origin_counts.csv')
df.head()
```

```
top_10_origins = df.sort_values(by='Count', ascending=False).head(10)
print(top_10_origins)
```

```
def decompose_categorical(x):
    if x=='Việt Nam':
        return 'Việt Nam'
    elif x=='Trung Quốc':
        return 'Trung Quốc'
    elif x=='China':
        return 'Trung Quốc'
    elif x=='china':
        return 'Trung Quốc'
    elif x=='Hàn Quốc':
        return 'Hàn Quốc'
    elif x=='Nhật Bản':
        return 'Nhật Bản'
    elif x=='Đài Loan':
        return 'Đài Loan'
    elif x=='Mỹ':
        return 'Mỹ'
    else:
        return 'Others'
    
df['Origin'] = df['Origin'].apply(decompose_categorical)
```

```

# Group by 'origin' and sum the 'count' for each group
grouped = df.groupby('Origin')['Count'].sum().reset_index()
```

```

# Sort the result in descending order
sorted_result = grouped.sort_values(by='Count', ascending=False)

# Print the sorted result
print(sorted_result)
```

```

# Create a pie chart
plt.figure(figsize=(16, 8))
plt.pie(grouped['Count'], labels=grouped['Origin'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Top 6 Origins by Count')

# Show the pie chart
plt.show()

# Save the pie chart as an image file (e.g., PNG)
plt.savefig('pie_chart.png')
```
